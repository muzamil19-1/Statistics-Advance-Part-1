**1.What is a random variable in probability theory.**

Ans:-In probability theory, a random variable is a variable that represents the outcome of a random experiment.

üîπ Definition:

A random variable is a function that assigns a real number to each outcome in the sample space of a random process.

**2.What are the types of random variables.**

Ans:-There are two main types of random variables in probability theory:

üîπ 1. Discrete Random Variable
Definition: Takes on a finite or countably infinite set of distinct values.

Values: Often integers or specific categories.

Examples:

Number of heads when flipping 3 coins ‚Üí {0, 1, 2, 3}

Number of students in a class

Rolling a die ‚Üí {1, 2, 3, 4, 5, 6}

‚úÖ Key Features:

Values are separated (not continuous)

Can be described using a probability mass function (PMF)

üîπ 2. Continuous Random Variable

Definition: Takes on uncountably infinite values, typically from an interval or range of real numbers.

Values: Can be any number within a range (including decimals).

Examples:

Height of a person

Time to complete a task (e.g., 3.5 sec, 3.51 sec, 3.512... sec)

Temperature on a given day

**3.What is the difference between discrete and continuous distributions.**

Ans:-The difference between discrete and continuous distributions lies in the type of values the random variable can take and how probabilities are assigned.

üîπ 1. Discrete Distribution

Describes the probability distribution of a discrete random variable.

The values are countable and often integers.

Probabilities are assigned to specific values.

| Feature                    | Description                     |
| -------------------------- | ------------------------------- |
| Values taken               | Finite or countably infinite    |
| Example values             | 0, 1, 2, 3, ‚Ä¶                   |
| Probability at exact value | Non-zero (e.g., P(X = 2) = 0.3) |
| Total probability          | Sum of all probabilities = 1    |
| Graph type                 | Bar graph                       |
| Function used              | PMF (Probability Mass Function) |

 Examples:
Tossing a coin: number of heads

Rolling a die: outcome 1 to 6

Number of calls received in an hour

üîπ 2. Continuous Distribution

Describes the probability distribution of a continuous random variable.

The values are uncountably infinite, usually over an interval.

Probability of a single point is zero, so we measure probability over intervals.

‚úÖ Characteristics:

| Feature                    | Description                          |
| -------------------------- | ------------------------------------ |
| Values taken               | Infinite and uncountable             |
| Example values             | Any real number like 3.5, 4.72, etc. |
| Probability at exact value | Always zero (P(X = x) = 0)           |
| Total probability          | Area under curve = 1                 |
| Graph type                 | Smooth curve                         |
| Function used              | PDF (Probability Density Function)   |

üî∏ Examples:
Height or weight of people

Time taken to complete a task

Temperature readings

**4.What are probability distribution functions (PDF).**

Ans:-A Probability Density Function (PDF) is a function that describes the relative likelihood for a continuous random variable to take on a particular value.

üîπ Key Properties of a PDF:

| Property                                | Description                                          |
| --------------------------------------- | ---------------------------------------------------- |
| $f(x) \geq 0$ for all $x$               | PDF is always non-negative                           |
| The **total area under the curve** = 1  | Ensures the total probability is 1                   |
| $P(a \leq X \leq b) = \int_a^b f(x) dx$ | Probability is calculated over intervals, not points |
| $P(X = x) = 0$                          | Probability at any exact value is 0                  |

**5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF).**

Ans:-Great question! The Cumulative Distribution Function (CDF) and the Probability Density Function (PDF) are both used to describe probability distributions, but they serve different purposes and behave differently.

üîπ Difference Between PDF and CDF

| Feature                          | **PDF** (Probability Density Function)           | **CDF** (Cumulative Distribution Function)                    |
| -------------------------------- | ------------------------------------------------ | ------------------------------------------------------------- |
| **Type of variable**             | Continuous (PDF is for continuous variables)     | Both continuous and discrete variables                        |
| **Definition**                   | Describes **how likely** a value is near a point | Describes the **probability** that a value is ‚â§ a given value |
| **Mathematical Form**            | $f(x)$                                           | $F(x) = P(X \leq x)$                                          |
| **Probability over an interval** | $P(a \leq X \leq b) = \int_a^b f(x)\,dx$         | $P(a \leq X \leq b) = F(b) - F(a)$                            |
| **Value at a point**             | $P(X = x) = 0$ (for continuous)                  | $F(x)$ gives total probability up to $x$                      |
| **Graph**                        | Curve showing density (not total probability)    | Curve that always increases from 0 to 1                       |
| **Range**                        | $f(x) \geq 0$; no upper limit on value           | $0 \leq F(x) \leq 1$                                          |

Ans:-üìò Assumptions for Applying the Central Limit Theorem (CLT)
The Central Limit Theorem (CLT) is a powerful concept, but it has a few important assumptions and conditions to ensure it works properly.

‚úÖ 1. Independent and Identically Distributed (i.i.d.) Samples
The sample observations should be independent (one value doesn't affect another).

All observations should come from the same population (identically distributed).

üîπ Why?
This ensures that each sample contributes equally and fairly to the overall distribution.

**3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
        x (int): Outcome (0 or 1)
        p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
        float: Probability of observing x
    """
    if x not in [0, 1]:
        return 0.0
    return p if x == 1 else 1 - p
```

**9.Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores from a dataset and plots them.

    Parameters:
        data (list or np.array): Input numerical dataset
    """
    # Convert to numpy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.plot(z_scores, marker='o', linestyle='-', color='teal')
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.axhline(2, color='red', linestyle='--', linewidth=1, label='Z = 2')
    plt.axhline(-2, color='red', linestyle='--', linewidth=1, label='Z = -2')
    plt.title("Z-Scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores
```

```python
data = [10, 12, 13, 9, 15, 18, 14, 11, 16, 20]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

**15.Implement hypothesis testing using Z-statistics for a sample dataset**

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([52, 49, 51, 53, 50, 48, 54, 55, 49, 51])

# Known population parameters
mu_0 = 50             # Hypothesized population mean (H0: Œº = 50)
sigma = 3             # Known population standard deviation
alpha = 0.05          # Significance level

# Step 1: Calculate sample mean and size
sample_mean = np.mean(sample)
n = len(sample)

# Step 2: Calculate the Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 3: Calculate the critical Z-value for a two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Step 4: Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 5: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z:.2f}")
print(f"Critical Z-Value: ¬±{z_critical:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Conclusion
if abs(z) > z_critical:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

**16.Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.array([52, 49, 51, 53, 50, 48, 54, 55, 49, 51])

# Parameters
confidence_level = 0.95
mean = np.mean(data)
std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Standard error
se = std / np.sqrt(n)

# Z-critical value for 95% confidence
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)

# Confidence interval
margin_of_error = z_critical * se
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Output
print(f"Sample Mean: {mean:.2f}")
print(f"Confidence Interval (95%): ({lower_bound:.2f}, {upper_bound:.2f})")
```

**24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    """
    Generates a normal dataset and calculates skewness and kurtosis.

    Parameters:
        mean (float): Mean of the normal distribution
        std (float): Standard deviation of the distribution
        size (int): Number of data points

    Returns:
        Tuple: (skewness, kurtosis)
    """
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # By default, Fisher's definition (normal = 0)

    # Display results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Fisher's): {kurt:.4f}")

    return skewness, kurt

# Example usage
analyze_normal_distribution_skew_kurtosis()
```
